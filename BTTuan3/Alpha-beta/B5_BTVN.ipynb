```python
import math

states = {
    "root": ["A", "B", "C"],

    "A": ["A1", "A2"],
    "B": ["B1", "B2"],
    "C": ["C1", "C2"],

    # lá → điểm đánh giá
    "A1": 3,
    "A2": 5,
    "B1": 2,
    "B2": 9,
    "C1": 1,
    "C2": 4
}

# --------------------------------------------
# ALPHA-BETA
# --------------------------------------------
def alphabeta(node, alpha, beta, maximizing):
    if isinstance(states[node], int):  # node lá
        return states[node]

    if maximizing:
        value = -math.inf
        for child in states[node]:
            value = max(value, alphabeta(child, alpha, beta, False))
            alpha = max(alpha, value)
            if alpha >= beta:
                break
        return value
    else:
        value = math.inf
        for child in states[node]:
            value = min(value, alphabeta(child, alpha, beta, True))
            beta = min(beta, value)
            if alpha >= beta:
                break
        return value

# --------------------------------------------
# CHẠY THỬ
# --------------------------------------------
if __name__ == "__main__":
    print("Ứng dụng Alpha-Beta trong cờ vua / cờ tướng / cờ vây.")
    best = alphabeta("root", -math.inf, math.inf, True)
    print("Giá trị tốt nhất AI đạt được là:", best)
```
