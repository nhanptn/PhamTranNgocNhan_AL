```python
import math
import copy

X = "X"
O = "O"
EMPTY = None

# --------------------------------------------
# CẤU HÌNH BOARD
# --------------------------------------------
N = 5          # kích thước bàn cờ
WIN_LEN = 4    # số quân liên tiếp để thắng

# --------------------------------------------
# KHỞI TẠO BOARD
# --------------------------------------------
def initial_state():
    return [[EMPTY for _ in range(N)] for _ in range(N)]

# --------------------------------------------
# NGƯỜI ĐI HIỆN TẠI
# --------------------------------------------
def player(board):
    count = sum(row.count(X) + row.count(O) for row in board)
    return X if count % 2 == 0 else O

# --------------------------------------------
# DANH SÁCH NƯỚC ĐI
# --------------------------------------------
def actions(board):
    moves = []
    for i in range(N):
        for j in range(N):
            if board[i][j] == EMPTY:
                moves.append((i, j))
    return moves

# --------------------------------------------
# TRẠNG THÁI SAU KHI ĐI
# --------------------------------------------
def result(board, action):
    (i, j) = action
    b = copy.deepcopy(board)
    b[i][j] = player(board)
    return b

# --------------------------------------------
# KIỂM TRA THẮNG (WIN_LEN)
# --------------------------------------------
def check_line(line):
    cntX = cntO = 0
    for c in line:
        if c == X:
            cntX += 1; cntO = 0
            if cntX >= WIN_LEN: return X
        elif c == O:
            cntO += 1; cntX = 0
            if cntO >= WIN_LEN: return O
        else:
            cntX = cntO = 0
    return None

def winner(board):
    # hàng
    for i in range(N):
        w = check_line(board[i])
        if w: return w

    # cột
    for j in range(N):
        col = [board[i][j] for i in range(N)]
        w = check_line(col)
        if w: return w

    # chéo chính
    for base in range(N):
        d1 = []; d2 = []
        for k in range(N):
            if 0 <= base+k < N:
                d1.append(board[k][base+k])
                d2.append(board[base+k][k])
        if len(d1) >= WIN_LEN and check_line(d1): return check_line(d1)
        if len(d2) >= WIN_LEN and check_line(d2): return check_line(d2)

    # chéo phụ
    for base in range(N):
        d3 = []; d4 = []
        for k in range(N):
            if 0 <= base-k < N:
                d3.append(board[k][base-k])
                d4.append(board[base-k][k])
        if len(d3) >= WIN_LEN and check_line(d3): return check_line(d3)
        if len(d4) >= WIN_LEN and check_line(d4): return check_line(d4)

    return None

# --------------------------------------------
# KẾT THÚC GAME?
# --------------------------------------------
def terminal(board):
    if winner(board): return True
    for i in range(N):
        for j in range(N):
            if board[i][j] == EMPTY:
                return False
    return True

# --------------------------------------------
# UTILITY
# --------------------------------------------
def utility(board):
    w = winner(board)
    if w == X: return 1
    if w == O: return -1
    return 0

# --------------------------------------------
# ALPHA - BETA
# --------------------------------------------
def max_value(board, depth, alpha, beta):
    if depth == 0 or terminal(board):
        return utility(board)
    v = -math.inf
    for action in actions(board):
        v = max(v, min_value(result(board, action), depth-1, alpha, beta))
        alpha = max(alpha, v)
        if alpha >= beta:
            break
    return v

def min_value(board, depth, alpha, beta):
    if depth == 0 or terminal(board):
        return utility(board)
    v = math.inf
    for action in actions(board):
        v = min(v, max_value(result(board, action), depth-1, alpha, beta))
        beta = min(beta, v)
        if alpha >= beta:
            break
    return v

def alphabeta(board, depth=3):
    turn = player(board)
    best_move = None

    if turn == X:
        best = -math.inf
        for action in actions(board):
            v = min_value(result(board, action), depth-1, -math.inf, math.inf)
            if v > best:
                best = v
                best_move = action
    else:
        best = math.inf
        for action in actions(board):
            v = max_value(result(board, action), depth-1, -math.inf, math.inf)
            if v < best:
                best = v
                best_move = action
    return best_move

# --------------------------------------------
# CHẠY THỬ
# --------------------------------------------
if __name__ == "__main__":
    board = initial_state()
    print(f"Chọn X hoặc O: ")
    user = input().upper()

    while True:
        for r in board: print(r)

        if terminal(board):
            w = winner(board)
            print("Kết thúc! Kết quả:", w if w else "Hòa")
            break

        if player(board) == user:
            i = int(input("row = "))
            j = int(input("col = "))
            if board[i][j] == EMPTY:
                board[i][j] = user
        else:
            print("AI đang suy nghĩ (Alpha-Beta)…")
            move = alphabeta(board, depth=3)
            board = result(board, move)
```
