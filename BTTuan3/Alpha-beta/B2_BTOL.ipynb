```python

# BÀI 2 – THUẬT TOÁN ALPHA–BETA PRUNING CHO TICTACTOE 3x3
# File: bai2_alphabeta_3x3.py
# ======================================================

import copy
import math

X = "X"
O = "O"
EMPTY = None

# --------------------------------------------
# TRẠNG THÁI BAN ĐẦU
# --------------------------------------------
def initial_state():
    return [
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY]
    ]

# --------------------------------------------
# XÁC ĐỊNH NGƯỜI CHƠI HIỆN TẠI
# --------------------------------------------
def player(board):
    count = sum(row.count(X) + row.count(O) for row in board)
    return X if count % 2 == 0 else O

# --------------------------------------------
# TẬP HÀNH ĐỘNG CÓ THỂ ĐI
# --------------------------------------------
def actions(board):
    return {(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY}

# --------------------------------------------
# TRẠNG THÁI SAU KHI ĐI (i,j)
# --------------------------------------------
def result(board, action):
    (i, j) = action
    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)
    return new_board

# --------------------------------------------
# KIỂM TRA NGƯỜI THẮNG
# --------------------------------------------
def winner(board):
    # hàng
    for row in board:
        if row[0] == row[1] == row[2] != EMPTY:
            return row[0]

    # cột
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] != EMPTY:
            return board[0][col]

    # chéo
    if board[0][0] == board[1][1] == board[2][2] != EMPTY:
        return board[0][0]

    if board[0][2] == board[1][1] == board[2][0] != EMPTY:
        return board[0][2]

    return None

# --------------------------------------------
# KẾT THÚC GAME?
# --------------------------------------------
def terminal(board):
    if winner(board) is not None:
        return True
    return all(board[i][j] != EMPTY for i in range(3) for j in range(3))

# --------------------------------------------
# GIÁ TRỊ LỢI ÍCH
# X thắng: 1
# O thắng: -1
# Hòa: 0
# --------------------------------------------
def utility(board):
    w = winner(board)
    if w == X:
        return 1
    if w == O:
        return -1
    return 0

# --------------------------------------------
# MAX (cho X) – có alpha-beta
# --------------------------------------------
def max_value(board, alpha, beta):
    if terminal(board):
        return utility(board)

    v = -math.inf
    for action in actions(board):
        v = max(v, min_value(result(board, action), alpha, beta))
        alpha = max(alpha, v)
        if alpha >= beta:
            break  # cắt nhánh
    return v

# --------------------------------------------
# MIN (cho O) – có alpha-beta
# --------------------------------------------
def min_value(board, alpha, beta):
    if terminal(board):
        return utility(board)

    v = math.inf
    for action in actions(board):
        v = min(v, max_value(result(board, action), alpha, beta))
        beta = min(beta, v)
        if alpha >= beta:
            break  # cắt nhánh
    return v

# --------------------------------------------
# HÀM ALPHA–BETA TRẢ VỀ NƯỚC ĐI TỐI ƯU
# --------------------------------------------
def alphabeta(board):
    turn = player(board)

    if turn == X:
        best = -math.inf
        best_move = None
        alpha = -math.inf
        beta = math.inf

        for action in actions(board):
            value = min_value(result(board, action), alpha, beta)
            if value > best:
                best = value
                best_move = action
            alpha = max(alpha, best)
        return best_move

    else:  # O
        best = math.inf
        best_move = None
        alpha = -math.inf
        beta = math.inf

        for action in actions(board):
            value = max_value(result(board, action), alpha, beta)
            if value < best:
                best = value
                best_move = action
            beta = min(beta, best)
        return best_move

# --------------------------------------------
# CHẠY THỬ
# --------------------------------------------
if __name__ == "__main__":
    board = initial_state()
    print("Bạn chọn X hay O?")
    user = input().upper()
    ai = O if user == X else X

    while True:
        print("\nBoard:")
        for r in board:
            print(r)

        if terminal(board):
            w = winner(board)
            if w:
                print(f"\nGame Over! {w} thắng.")
            else:
                print("\nHòa!")
            break

        if player(board) == user:
            print("\nNhập nước đi của bạn (row col): ")
            i = int(input("row = "))
            j = int(input("col = "))
            if board[i][j] == EMPTY:
                board[i][j] = user
        else:
            print("\nAI đang suy nghĩ (Alpha–Beta)…")
            move = alphabeta(board)
            board = result(board, move)
```
